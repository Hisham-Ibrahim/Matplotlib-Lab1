## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(df['Fare'],df['SibSp'], marker = 'o', s = 50, alpha=0.4,label = 'Fare') # s >> size

plt.legend(loc = 'upper center')
plt.title('Fare vs. SibSp',fontsize = 20)
plt.xlabel('Fare',fontsize = 20)
plt.ylabel('SibSp',fontsize = 20)

plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python

plt.scatter(df['Fare'],df['SibSp'], marker = 'o', s = 50, alpha=0.4,label = 'Fare') # s >> size
plt.legend(loc = 'upper center')
font1 = {'color':'red','size':16}
plt.title("Fare vs. SibSp", fontdict = font1,loc = 'left')
plt.xlabel('Fare',fontsize = 16)
plt.ylabel('SibSp',fontsize = 16)

plt.show()
```
